```python
import numpy as np
dict_values ={100:{}, 150:{}, 200:{}, 500:{}, 800:{}, 1000:{}}
for key in dict_values.keys():
    dict_values[key] = {0.25:{}, 1.2:{}, 5:{}}
threshold = 0.025
output_signal = coffea.util.load( "output_signal_base_ljObjCut_ljIso_2lj.coffea" )
for s in signals_all:
    parts = s.split("_")
    
    histogram = output_signal["out"][s]["hists"]["mu_lj_pfMuon_min_dxy_XLowRange"]["base_ljObjCut_ljIso_2lj", :]
    values = histogram.values()
    overflow = histogram.values(flow=True)[-1]
    # print(overflow)
    bin_edges = histogram.axes[0].edges
    idx = np.searchsorted(bin_edges, threshold, side="right")
    n_events = values[idx:].sum() + overflow
    n_events_total = values.sum() + overflow
    n_fraction = n_events/n_events_total
    # utilities.plot(histogram, density=True, yerr=False)
    mass = float(parts[1].replace("GeV", ""))
    zd_mass = float(parts[2].replace("p", ".").replace("GeV", ""))
    ctau = float(parts[3].replace("p", ".").replace("mm", ""))

    result = (mass, zd_mass, ctau)
    # print(result)
    dict_values[mass][zd_mass].update( { ctau: [ n_events, n_fraction]})
# print(dict_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

def plot_signal_and_percentage_grids(data):
    for mass, zd_dict in data.items():
        zd_vals = sorted(zd_dict.keys())
        lab_vals = sorted(set(lab for zd in zd_vals for lab in zd_dict[zd].keys()))

        # Initialize grids
        z_yield = np.full((len(lab_vals), len(zd_vals)), np.nan)
        z_percent = np.full((len(lab_vals), len(zd_vals)), np.nan)

        for i, lab in enumerate(lab_vals):
            for j, zd in enumerate(zd_vals):
                values = zd_dict.get(zd, {}).get(lab, [np.nan, np.nan])
                z_yield[i, j] = values[0]
                z_percent[i, j] = values[1]

        fig, axs = plt.subplots(1, 2, figsize=(24, 10), constrained_layout=True)

        vmin_fixed = 1e-2   # smallest meaningful yield (e.g., 0.01)
        vmax_fixed = 1e2 
        # --- Plot yield ---
        im1 = axs[0].imshow(
            z_yield,
            origin='lower',
            cmap='viridis',
            norm=LogNorm(vmin=vmin_fixed, vmax=vmax_fixed),
            aspect='equal'
        )
        axs[0].set_xticks(np.arange(len(zd_vals)))
        axs[0].set_xticklabels(zd_vals)
        axs[0].set_yticks(np.arange(len(lab_vals)))
        axs[0].set_yticklabels(lab_vals)
        axs[0].set_title(f"Signal Yield (mass = {mass} GeV)")
        axs[0].set_xlabel(r"$m_{Z_d}$ (GeV)")
        axs[0].set_ylabel(r"$L_{xy}$ (cm)")
        plt.colorbar(im1, ax=axs[0], label="Events (log scale)")

        # --- Plot percentage ---
        im2 = axs[1].imshow(
            z_percent,
            origin='lower',
            cmap='viridis',
            vmin=0, vmax=1,  # Since percentages are in range 01
            aspect='equal'
        )
        axs[1].set_xticks(np.arange(len(zd_vals)))
        axs[1].set_xticklabels(zd_vals)
        axs[1].set_yticks(np.arange(len(lab_vals)))
        axs[1].set_yticklabels(lab_vals)
        #axs[1].set_title(f"Fraction of Total Events (mass = {mass} GeV)")
        axs[1].set_xlabel(r"$m_{Z_d}$ (GeV)")
        axs[1].set_ylabel(r"$L_{xy}$ (cm)")
        plt.colorbar(im2, ax=axs[1], label=r"$Fraction$")

        # plt.suptitle(f"Signal Grid for Mass = {mass} GeV", fontsize=16)
        plt.show()
        plt.close()

plot_signal_and_percentage_grids(dict_values_lxy)
```

```python
def plot_percentage_grid_all(data):
    import numpy as np
    import matplotlib.pyplot as plt

    # Collect all (bound_mass, zd_mass) pairs
    combos = []
    for mass, zd_dict in data.items():
        for zd in zd_dict.keys():
            combos.append((mass, zd))
    combos = sorted(combos, key=lambda x: (x[0], x[1]))

    # Collect all Lxy values (y-axis)
    lab_vals = sorted(set(
        lab for mass in data for zd in data[mass] for lab in data[mass][zd].keys()
    ))

    # Initialize grid for percentage
    z_percent = np.full((len(lab_vals), len(combos)), np.nan)

    # Fill grid
    for j, (mass, zd) in enumerate(combos):
        for i, lab in enumerate(lab_vals):
            values = data[mass][zd].get(lab, [np.nan, np.nan])
            z_percent[i, j] = values[1]

    # X-axis labels like "100/0.25"
    x_labels = [f"{m}/{zd}" for m, zd in combos]

    # --- Plot percentage ---
    fig, ax = plt.subplots(figsize=(15, 8), constrained_layout=True)
    im = ax.imshow(
        z_percent,
        origin='lower',
        cmap='viridis',
        vmin=0, vmax=1,
        aspect='auto'
    )

    ax.set_xticks(np.arange(len(combos)))
    ax.set_xticklabels(x_labels, rotation=90)
    ax.set_yticks(np.arange(len(lab_vals)))
    ax.set_yticklabels(lab_vals)
    # ax.set_title(r"$d_{xy} > 0.025$ cm")
    ax.set_xlabel(r"Bound state / $m_{Z_d}$ (GeV)")
    ax.set_ylabel(r"$L_{xy}$ (cm)")
    plt.colorbar(im, ax=ax, label=r"Fraction ($d_{xy}>$ 0.025 cm)" )
    hep.cms.label()

    plt.show()
    plt.close()

plot_percentage_grid_all(dict_values_lxy)
```
